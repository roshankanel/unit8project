```python
BTC = yf.Ticker("BTC-USD")
```

```python
BTC = BTC.history(period='max')
```

```python
del BTC['Dividends']
del BTC['Stock Splits']       
```

```python
BTC['Tomorrow'] = BTC['Close'].shift(-1)
```

```python
BTC
```

```python
BTC['Target'] = (BTC['Tomorrow'] > BTC['Close']).astype(int)
```

```python
BTC
```

```python
BTC = BTC.loc['1990-1-1':].copy()
```

```python
BTC
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)

train = BTC.iloc[:-100]
test = BTC.iloc[-100:]

predictors = ['Close', 'Volume', 'Open', 'High', 'Low']
model.fit(train[predictors], train['Target'])
```

```python
from sklearn.metrics import precision_score
preds = model.predict(test[predictors])
```

```python
import pandas as pd
preds = pd.Series(preds, index=test.index)
```

```python
precision_score(test ['Target'], preds)
```

```python
combined = pd.concat([test['Target'], preds], axis = 1)
combined.plot()
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train['Target'])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index = test.index, name='Predictions')
    combined = pd.concat([test['Target'], preds], axis = 1)
    return combined
```

```python
def backtest(data, model, predictors, start = 2500, step = 250):
    all_predictions = []
    
    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
        return pd.concat(all_predictions)
```

```python
predictions = backtest(BTC, model, predictors)
print(predictions)
```

```python
predictions['Predictions'].value_counts()
```

```python
precision_score(predictions['Target'], predictions['Predictions'])
```

```python
# Additing Additional Predictors to improve accuracy
# Rolling  averages
```

```python
horizons = [2,5,60,250,1000]
new_predictors = []

for horizon in horizons:
    rolling_averages = BTC.rolling(horizon).mean()
    
    ratio_column = f"Close_Ratio_{horizon}"
    BTC[ratio_column] = BTC['Close'] / rolling_averages['Close']
    
    trend_column = f"Trend_{horizon}"
    BTC[trend_column] = BTC.shift(1).rolling(horizon).sum()["Target"]
    
    new_predictors += [ratio_column, trend_column]
```

```python
#improving model
#increasing accuracy by increasing estimators and decreaskig split
# addding predict_proba will retun a probability not just a 1 or 0, probability it will go up or probbility it will go down.
```

```python
model = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state = 1)
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train['Target'])
    preds = model.predict_proba(test[predictors])[:,1]
    preds[preds >= .6] = 1
    preds[preds < .6] = 0
    preds = pd.Series(preds, index = test.index, name='Predictions')
    combined = pd.concat([test['Target'], preds], axis = 1)
    return combined
```

```python
predictions = backtest(BTC, model, new_predictors)
print(new_predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```
